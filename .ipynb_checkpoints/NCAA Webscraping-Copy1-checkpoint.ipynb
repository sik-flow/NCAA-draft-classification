```python
#Import moduels
import requests
from bs4 import BeautifulSoup as bs
```

```python
#URL

URL = "https://en.wikipedia.org/wiki/List_of_NCAA_Division_I_men's_basketball_programs"
page = requests.get(URL)
```

```python
# Print html content of the page
print(page.content)
```

```python
web = bs(page.content,'html.parser')
```

```python
print(web)
```

```python
#Finding Elements by class "wikitable sortable jquery-tablesorter"

wikitable = web.find('table',{'class':'wikitable sortable'})
```

```python
print(wikitable)
```

```python
#find all table row
wikitable.find_all('tr')[1:]
```

```python
#get all names of D1 colleges
college_names = []
for row in wikitable.find_all('tr')[1:]:
    college_names.append(row.find('a').text)
```

```python
temp = []
for row in wikitable.find_all('tr')[1:]:
    for x in row.find_all('td'):
        temp.append(x.text.splitlines())
```

```python
temp
```

```python
for ele in temp[::7]:
    print(ele)
    
```
